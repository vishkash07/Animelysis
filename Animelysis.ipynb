###Data Cleaning <a name="data-cleaning"></a>
Once we have identified our data and the inconsistencies it contains, we can filter out or clean those discrepancies to ensure that we have meaningful and reliable data for analysis

```python
df.isnull().sum()
```

```python
# List of columns to remove
columns_to_remove = ['Watching', 'Favorites', 'Completed', 'On-Hold', 'Dropped', 'Plan to Watch']
# Drop the columns
df_anime = df.drop(columns=columns_to_remove)
df_anime.head(10).T
```

```python
columns_to_remove = [f"Score-{i}" for i in range(1, 11)]
# Drop the specified columns
df_anime.drop(columns=columns_to_remove, inplace=True)
```

```python
for column in df_anime.columns:
  print(f"Column: {column}")
  print(df_anime[column].value_counts())
  print("\n")
```

*Cleaning "Type" Column*

```python
df_anime = df_anime.loc[df_anime['Type'] != 'Unknown']
df_anime['Type'].value_counts()
```

*Cleaning "Score" Column*

```python
df_anime['Score'] = pd.to_numeric(df_anime['Score'], errors='coerce')
mean_score = df_anime['Score'].mean()
df_anime['Score'].fillna(mean_score, inplace=True)
df_anime.loc[df_anime['Score'].isna()]
```

*Creating a new column "Duration_Minutes" to convert the Duration values to Minutes*

```python
def duration_to_minutes(duration_str):
    parts = duration_str.split()
    total_minutes = 0
    for i in range(len(parts)):
        if parts[i].isdigit():
            if parts[i + 1] == 'hr.':
                total_minutes += int(parts[i]) * 60
            elif parts[i + 1] == 'min.':
                total_minutes += int(parts[i])
            elif parts[i + 1] == 'sec.':
                total_minutes += int(parts[i]) / 60
    return total_minutes
df_anime['Duration_Minutes'] = df_anime['Duration'].apply(duration_to_minutes)
```

###Analysis of Features <a name="analysis-of-features"></a>
In this section, we'll delve deeper into the dataset by visualizing its features. Through the use of plots and charts, we aim to gain insights into the distribution, relationships, and patterns within the data.

```python
anime_count_by_type = df_anime['Type'].value_counts()

# Plotting
plt.figure(figsize=(8, 5))
anime_count_by_type.plot(kind='bar', color=plt.cm.tab20c.colors)

# Adding title and labels
plt.title('Count of Anime Titles by Type')
plt.xlabel('Anime Type')
plt.ylabel('Count')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Sort the DataFrame by the number of users who have scored the anime
top_15_scored = df_anime.sort_values(by='Members', ascending=False).head(15)

# Convert the number of users to millions
top_15_scored['Members_Millions'] = top_15_scored['Members'] / 1_000_000  # Divide by 1 million

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(top_15_scored['Name'], top_15_scored['Members_Millions'], color=plt.cm.tab10.colors)  # Use a single color for all bars

# Adding title and labels
plt.title('Top 15 Animes by Number of Users')
plt.xlabel('Anime Title')
plt.ylabel('Number of Users (Millions)')  # Update ylabel
plt.yticks(rotation=45)  # Rotate y-axis labels

# Rotating x-axis labels for better readability
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

###Corelation Analysis <a name="correlation-analysis"></a>
In this section, we will explore the relationships between different features of the anime dataset. We will calculate correlation coefficients to quantify the degree of linear relationship between pairs of numerical variables.

This analysis will help us understand how various attributes of anime titles are related to each other and identify any potential patterns or trends.

```python
df_anime.info()
```

```python
scores_scatter = df_anime['Score']
durations_scatter = df_anime['Duration_Minutes']

# Plotting the data
plt.figure(figsize=(8, 6))
plt.scatter(durations_scatter, scores_scatter, color='blue', alpha=0.5)
plt.title('Score vs Duration (Minutes) of Anime')
plt.xlabel('Duration (Minutes)')
plt.ylabel('Score')
plt.grid(True)
plt.show()
```

```python
df_anime['Episodes'] = pd.to_numeric(df_anime['Episodes'], errors='coerce')
```

```python
df_anime.loc[(df_anime['Episodes']> 2000) & (df_anime['Episodes'] < 3000)]
```

```python
bins = [0, 500, 1000, 1500, 2000, 2500, 3000, float('inf')]

df_anime['Episodes_Bins'] = pd.cut(df_anime['Episodes'], bins=bins, labels=['0-499', '500-999', '1000-1499', '1500-1999', '2000-2499', '2500-2999', '3000+'])

average_scores = df_anime.groupby('Episodes_Bins')['Score'].mean()
plt.figure(figsize=(10, 6))
average_scores.plot(kind='bar', color='skyblue')
plt.title('Average Score by Episodes Range')
plt.xlabel('Episodes Range')
plt.ylabel('Average Score')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
ranked_scatter = df_anime[~df_anime['Ranked'].isna()]
plt.scatter(ranked_scatter['Ranked'], ranked_scatter['Score'], color='red', alpha=0.5)
plt.title('Score vs Ranked of Anime')
plt.xlabel('Ranked')
plt.ylabel('Score')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

numerical_variables = ['Score', 'Duration_Minutes', 'Popularity']

correlation_matrix = df_anime[numerical_variables].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap: Score, Duration, Popularity')
plt.show()
```
